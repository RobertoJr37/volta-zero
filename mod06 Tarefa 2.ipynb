```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o arquivo CSV
df = pd.read_csv('SINASC_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Filtrando a coluna IDADEMAE, removendo valores nulos
df_filtered = df[['IDADEMAE']].dropna()

# Criando o boxplot para IDADEMAE
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_filtered['IDADEMAE'])
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Filtrando a coluna PESO, removendo valores nulos
df_filtered = df[['PESO']].dropna()

# Criando o boxplot para PESO
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_filtered['PESO'])
plt.title('Boxplot do Peso ao Nascer')
plt.xlabel('Peso do Bebê (gramas)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Filtrando a coluna IDADEMAE, removendo valores nulos
df_filtered = df[['IDADEMAE']].dropna()

# Criando o histograma para IDADEMAE
plt.figure(figsize=(10, 6))
sns.histplot(df_filtered['IDADEMAE'], bins=20, kde=True)
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Filtrando a coluna PESO, removendo valores nulos
df_filtered = df[['PESO']].dropna()

# Criando o histograma para PESO
plt.figure(figsize=(10, 6))
sns.histplot(df_filtered['PESO'], bins=30, kde=True)
plt.title('Histograma do Peso ao Nascer do Bebê')
plt.xlabel('Peso do Bebê (gramas)')
plt.ylabel('Frequência')
plt.show()
```
